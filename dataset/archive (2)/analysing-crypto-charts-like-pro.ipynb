```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=list(btc.Date), y=list(btc.High)))

# Set title
fig.update_layout(
    title_text="BTC Time series with range slider and selectors"
)
fig.add_layout_image(
    dict(
        source="https://image.shutterstock.com/image-vector/bitcoin-logo-currency-crypto-here-260nw-767986060.jpg",
        xref="paper", yref="paper",
        x=1, y=1.05,
        sizex=0.2, sizey=0.2,
        xanchor="right", yanchor="bottom"
    )
)
# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.show()
```

```python
fig = go.Figure(data=[go.Candlestick(x=btc['Date'],
                open=btc['Open'],
                high=btc['High'],
                low=btc['Low'],
                close=btc['Close'])])

fig.add_layout_image(
    dict(
        source="https://image.shutterstock.com/image-vector/bitcoin-logo-currency-crypto-here-260nw-767986060.jpg",
        xref="paper", yref="paper",
        x=1, y=1.05,
        sizex=0.2, sizey=0.2,
        xanchor="right", yanchor="bottom"
    )
)

fig.update_layout(
    title='BTC Price Over Time',
    yaxis_title='BTC Price',
    shapes = [dict(
        x0='2021-11-01', x1='2022-07-20', y0=0, y1=1, xref='x', yref='paper',
        line_width=2)],
    annotations=[dict(
        x='2021-11-01', y=0.05, xref='x', yref='paper',
        showarrow=False, xanchor='right', text='Bear Market Begins')]
)

fig.show()
```
